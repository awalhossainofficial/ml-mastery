## Bar Chart
catagorical data show

```python

data = {
    'Student': [ 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Gender': [ 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'StudyHours': [ 2, 5, 3, 7, 4, 8, 1],
    'Attendance': [ 60, 88, 70, 95, 80, 98, 55],
    'Grade': [ 'Fail', 'Pass', 'Fail', 'Pass', 'Pass', 'Pass', 'Fail']
}

df = pd.DataFrame(data)
df
```

```python
 # Gender wise group
gender_group = df.groupby('Gender').size()
gender_group.index
gender_group.values
```

```python
plt.bar(gender_group.index, gender_group.values, color=["yellow","red"])
plt.xlabel("Gender")
plt.ylabel("Frequency of Student")
plt.title("Gender wise group")
plt.show()

```

 ## Bar Chart on Real Data

```python
# load csv data
df = pd.read_csv('student.csv')
df
```

```python
# student complettion status

status = df.groupby(df['CompletionStatus']).size()
status.index
status.values

plt.bar(status.index, status.values, color=['green', 'orange', 'red'], edgecolor='black')
plt.xlabel("Completion Status")
plt.ylabel("Count")
plt.title("Car chart of Course completion")
```

```python
# location 

status = df.groupby('Location').size() 

plt.bar(status.index,status.values,edgecolor='black')

plt.xlabel('Location') 
plt.ylabel('Count') 
plt.title('Bar Chart of Location of Students')
plt.show() 
```

## Pie Chart

```python
status = df.groupby('CompletionStatus').size()

plt.pie(status, labels=status.index, autopct='%1.1f%%', explode=(0,0.1,0.1), shadow=True)
plt.show()
```
