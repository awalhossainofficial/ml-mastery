### Part B – Percentiles, IQR and Z-Score

```python
weight = [2.5, 3.0, 3.2, 3.3, 3.4, 3.5, 3.6, 3.9, 4.0, 4.5]
wg = pd.DataFrame(weight,columns=['weight'])

q1 = wg['weight'].quantile(0.25)
q2 = wg['weight'].quantile(0.50)
q3 = wg['weight'].quantile(0.75)
iqr = q3 - q1
z_score = (65 - 50) / 10
x = -2.5 * 10 + 50

print(f"IQR: {iqr}")
print(f"P25 (Q1): {q1}")
print(f"P50 (Q2/Median): {q2}")
print(f"P75 (Q3): {q3}")

print(z_score,"z-score",x)
```
