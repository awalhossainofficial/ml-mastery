```python
with open('./email.txt') as f:
    mails = f.readlines()
    emails = [mail.strip().lower() for mail in mails]
    
    unique_mails = sorted(set(emails))
    strip = ("\n").join(unique_mails)
    # saving them in sorting way 
    with open('unique_emails.txt',"w") as file:
        file.write(strip)
            
```
