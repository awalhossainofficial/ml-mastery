```python
class Vehicle:
    def __init__(self, type, name, speed):
        self.__type = type
        self.__name = name
        self.speed = speed

    def classify(self):
        if self.speed > 100:
            return "High-speed Vehicle"
        else:
            return "Normal Vehicle"
    
    def show_info(self):
        return (f"Type {self.__type} Name {self.__name} speed: {self.speed} Predicted: {self.classify()}")


class Car(Vehicle):
    def show_info(self):
        return super().show_info()
    
class Bike(Vehicle):
    def show_info(self):
        return super().show_info()
```

```python
num_of_car = int(input("Enter number of the vehicle: "))
vehicles = []

for _ in range(num_of_car):
    vtype, name, speed = input("Enter <Type> <Name> <Speed>: ").split()
    speed = int(speed)
    if vtype.lower() == 'car':
       vehicles.append(Car(vtype, name, speed)) 
    if vtype.lower() == 'bike':
       vehicles.append(Bike(vtype, name, speed)) 

for v in vehicles:
   print(v.show_info())  # Call the method
```
