```python
try:
    inpt = input("Please provide the input: ").split()
    print(f"Input received: {inpt}")
    
    if not inpt:
        print("No input provided")
    else:
        even_int = list(filter(lambda a: a % 2 == 0, map(int, inpt)))
        print(f"Even numbers: {even_int}")
        
except ValueError as e:
    print(f"Please provide a right input - Error: {e}")
```
