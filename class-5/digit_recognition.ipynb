```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np  
```

```python
(X_train, y_train) , (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
X_train[0].shape
```

```python
X_train[0]
```

```python
plt.figure(figsize=(5,5))
plt.imshow(X_train[0], cmap='gray')
plt.title(f'Digit: {y_train[0]}')
plt.axis('off')
plt.show()

# plt.matshow(X_train[0])
```

```python
y_train[0]
```

```python
X_train = X_train / 255
X_test = X_test / 255
```

```python
X_train[0]
```

```python
X_train_flattened = X_train.reshape(len(X_train), 28*28)
X_test_flattened = X_test.reshape(len(X_test), 28*28)
```

```python
X_train_flattened[0]
```

***Very simple neural network with no hidden layers***

```python
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation='softmax') # softmax is used for classification
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy', # log loss
              metrics=['accuracy']
              )
model.fit(X_train_flattened, y_train, epochs=5)
```

```python
model.evaluate(X_test_flattened, y_test)
```

```python
y_predicted = model.predict(X_test_flattened)
y_predicted[0]
```

```python
# Example prediction
prediction = model.predict(X_test_flattened[0:1])
print("Model's output (probabilities):", prediction[0])
print("Highest probability digit:", np.argmax(prediction[0]))
print("Actual digit:", y_test[0])

# Example output:
# Model's output: [0.01, 0.02, 0.01, 0.01, 0.90, 0.01, 0.01, 0.01, 0.01, 0.01]
# Highest probability digit: 4
# Actual digit: 4
```

```python
# matplot show
plt.figure(figsize=(15,5))
plt.imshow(X_test[0], cmap='gray')
plt.title(f'Predicted: {np.argmax(prediction[0])}, Actual: {y_test[0]}')
plt.axis('off')
plt.show()
```

```python
y_predicted_lables = [np.argmax(i) for i in y_predicted]
```

```python
y_predicted_lables[:5]
```

```python
cm = tf.math.confusion_matrix(labels=y_test, predictions=y_predicted_lables)
cm
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.show()
```

**Using hidden layers**


```python
model = keras.Sequential([
    keras.layers.Dense(100, input_shape=(784,), activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy']
              )
model.fit(X_train_flattened, y_train, epochs=5)
```

```python
model.evaluate(X_test_flattened, y_test)
```

```python
y_predicted = model.predict(X_test_flattened)
y_predicted_lables = [np.argmax(i) for i in y_predicted]
cm = tf.math.confusion_matrix(labels=y_test, predictions=y_predicted_lables)

plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.show()

```

**Using Flatten layer so that we don't have to call reshape on input dataset**


```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy']
              )
model.fit(X_train, y_train, epochs=10)
```

```python
model.evaluate(X_test, y_test)
```

```python

y_predicted = model.predict(X_test)
y_predicted_lables = [np.argmax(i) for i in y_predicted]

cm = tf.math.confusion_matrix(labels=y_test, predictions=y_predicted_lables)

plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.show()
```
