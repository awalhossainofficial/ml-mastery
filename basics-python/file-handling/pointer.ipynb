```python
from functools import reduce


with open('./demo.txt') as file:
    # number of lines
    strings = file.readlines()
    print(len(strings), "lines", strings)

    # number of words
    words = list(map(lambda x: len(x.split()), strings))
    print(words)
    # total words 
    total_number_words = reduce(lambda acc, x: acc + x, words)

    # clean up the file like string 
    str_clean = list(map(lambda x: x.strip(), strings))
    print(str_clean)
    print(total_number_words,"word", words)
```
