```python
#1 load dataset 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('student_iq.csv')

# Scatter plot: Study Hours vs IQ Score
plt.xlabel("Study Hours")
plt.ylabel("Student IQ")
plt.title("Relationship between Study Hours and IQ Score")
plt.scatter(df['Study_Hour'], df['IQ_Score'])
plt.show()
```

```python
# Scatter plot: Shoe Size vs Study Hours
plt.xlabel("Shoe Size")
plt.ylabel("Study Hours")
plt.title("Relationship between Shoe Size and Study Hours")
plt.scatter(df['Shoe_Size'], df['Study_Hour'])
plt.show()
```

```python
# shoe 
plt.xlabel("Shoe Size")
plt.ylabel("Study Hours")

plt.scatter(df['Shoe_Size'], df['Study_Hour'])
```

## Analysis: Does there appear to be a correlation?

### 1. Study Hours vs IQ Score:
- **YES, there appears to be a positive correlation** 
- As study hours increase, IQ scores tend to increase
- However, the correlation is not very strong - there's considerable scatter in the data
- Some outliers exist (e.g., someone with low study hours but high IQ)

### 2. Shoe Size vs Study Hours:
- **NO clear correlation**
- Shoe size and study hours appear to be independent variables
- No obvious pattern or trend in the scatter plot
- This makes sense as shoe size (physical trait) shouldn't affect study habits

### 3. Shoe Size Distribution:
- The histogram shows that most students have shoe sizes between 6 and 13
- The distribution appears roughly normal with slight variation
