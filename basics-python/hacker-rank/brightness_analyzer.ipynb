```python
inpt = input().split()
total = 0
length_input = len(inpt)
for i in  inpt:
    total += int(i)
avg_brighness = total / length_input

if avg_brighness < 85:
    print("Dark Image")
elif 85 <= avg_brighness <=170:
    print("Normal Image")
elif avg_brighness > 170:
    print("Bright Image")

```
