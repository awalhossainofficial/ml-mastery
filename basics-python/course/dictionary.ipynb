```python
import string


text = """Python is a language every developer loves. A developer who writes Python code knows how clean and simple Python can be. The code written in Python is readable, reusable, and efficient. Every Python developer enjoys building projects, analyzing data, and creating automation scripts. Data analysis in Python makes a developer productive. Python handles data, text, and files with elegance. Many developer teams use Python to process data, train models, and clean data pipelines.

When a developer writes Python code, they focus on clarity. Python encourages developers to write less code that does more work. A developer working with data can use Python libraries like pandas, numpy, and matplotlib. These libraries make a developer’s life easier. Python and data go hand in hand. Clean data leads to better Python code, and better Python code helps manage complex data.

A good developer tests their Python code. Testing Python code ensures the code runs efficiently with data of all types. The developer who understands both Python and data can solve complex problems faster. When multiple developer teams collaborate, Python code becomes the shared language.

Whether it’s data science, web development, or automation, Python gives every developer a way to express logic cleanly. That’s why Python remains one of the most loved languages by every developer who works with code, with data, and with creativity."""

text = text.lower().translate(str.maketrans('','', string.punctuation))
extract = text.split()
dic = {}
# print(text,"chl")
for i in extract:
        dic[i] = dic.get(i,0)+1
for k,v in dic.items():
    print(f"{k}:{v}")
```
