```python
inpt = input("Please provide a number: ")

def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n-1)
try:
    total = 0
    for i in list(inpt): 
        print(i,"chel")
        total += factorial(int(i))

    if total == int(inpt):
        print("Strong Number")
    else:
        print("Not a Strong Number")
except(ValueError, TypeError):
    print("please provide a valid input ")
    # print(f"Total factorial sum: {total}")
```
