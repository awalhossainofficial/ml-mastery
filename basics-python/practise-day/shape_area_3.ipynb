```python
pi = 3.141592653589793  

class Shape:
    def area(self):
        return 0

class Rectangle(Shape):
    def __init__(self, w, h):
        self.w = w
        self.h = h

    def area(self):
        total = self.w * self.h
        return total

class Circle(Shape):
    def __init__(self, r):
        self.r = r
    
    def area(self):
        result = self.r**2 * pi
        return result
# input dynamically

try:
    inpt = input().split()
    shape_type = inpt[0]

    if shape_type.lower() == 'rectangle':
        w, h = map(float, inpt[1:3])
        shape = Rectangle(w,h)
        print(w,h)
    elif shape_type.lower() == 'circle':
        r = float(inpt[1])
        shape = Circle(r)
    else:
        raise ValueError("Unknow Shape type")
    
    print(f"Area: {shape.area():.2f}")


except (IndexError, ValueError):
    print("Error: Please provide valid input (e.g., 'Rectangle 5 10' or 'Circle 7')")
```
