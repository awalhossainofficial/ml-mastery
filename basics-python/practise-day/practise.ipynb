```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# load data
df = pd.read_csv('employees.csv')

df['Discounted_Price'] = df['Price'] * 0.9

df.to_csv('discounted_products.csv')
```

```python
# count duplicate
total_duplicate = df.duplicated().sum()
print(total_duplicate)

# removing duplicate - use .copy() to avoid SettingWithCopyWarning
rm_dup = df.drop_duplicates(subset=['EmployeeID']).copy()
rm_dup
```

```python
# handling missing values

# counting null values
null_counts = rm_dup.isnull().sum()
print(null_counts)

# Fill missing Age values with mean (assign back to update the dataframe)
rm_dup['Age'] = rm_dup['Age'].fillna(rm_dup['Age'].mean())
rm_dup

# Drop rows where salary is missing (either use inplace=True without assignment, or assign without inplace)
rm_dup.dropna(subset=['Salary'], inplace=True)
rm_dup
```

## Apply Statistical Functions


```python
# Task 1: Calculate mean and max for each subject
print("=== Mean for each subject ===")
print(f"Marks1 mean: {rm_dup['Marks1'].mean():.2f}")
print(f"Marks2 mean: {rm_dup['Marks2'].mean():.2f}")
print(f"Marks3 mean: {rm_dup['Marks3'].mean():.2f}")

print("\n=== Max for each subject ===")
print(f"Marks1 max: {rm_dup['Marks1'].max()}")
print(f"Marks2 max: {rm_dup['Marks2'].max()}")
print(f"Marks3 max: {rm_dup['Marks3'].max()}")

# Alternative: Using describe() for comprehensive statistics
print("\n=== Statistics for all marks columns ===")
print(rm_dup[['Marks1', 'Marks2', 'Marks3']].describe())

# Task 2: Compute total marks per student
rm_dup['Total_Marks'] = rm_dup['Marks1'] + rm_dup['Marks2'] + rm_dup['Marks3']
print("\n=== Total marks per student ===")
print(rm_dup[['EmployeeName', 'Marks1', 'Marks2', 'Marks3', 'Total_Marks']])

# Task 3: Find standard deviation of total marks
std_dev_total = rm_dup['Total_Marks'].std()
print(f"\n=== Standard Deviation of Total Marks ===")
print(f"Standard Deviation: {std_dev_total:.2f}")

rm_dup
```
