```python
class MobilePhone:
    def __init__(self, model, battery=100):
        self.model = model
        self.__battery = battery

    def use(self):
        if self.__battery > 0:
            self.__battery -= 1
    
    def charge(self):
        self.__battery = 100
    
    def get_battery(self):
        print(f"Battery percentage {self.__battery}")


myphone = MobilePhone("y",100)
print(myphone.model, "and charge")

myphone.get_battery()
myphone.use()
myphone.get_battery()

```
