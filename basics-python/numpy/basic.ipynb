## converting tuple, set, dic into ndarray


```python
import numpy as np
dic = {"a":1, "b":2, "c":3}
keys = dic.keys()
values = dic.values()

print(keys, "nl", values)
key_arr = np.array(list(dic.items()))
print(key_arr, key_arr.dtype, "check", key_arr.ndim,"size",key_arr.size)
```
