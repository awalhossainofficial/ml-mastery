```python
import numpy as np
a = np.random.randint(1,10, size=(2,3))
b = np.random.randint(20, 30, size=(1,3))
concatenate = np.concatenate((a,b),axis=0)
print("a\n",a)
print("b\n",b)
print("concate \n", concatenate)
```

## Transpose matrix

```python

arr = np.array([[10,20,30],
                [30,40]])

transpose = arr.T
print(transpose)
```
