### Q3. Confusion matrix and basic metrics

For this question, use:
- `true_label` as the actual label  
- `pred_label` as the model prediction


```python
# Q3.1: Manually compute TP, TN, FP, FN [We already write this ans]
true_col = "true_label"
pred_col = "pred_label"

tp = ((df[true_col] == 1) & (df[pred_col] == 1)).sum()
tn = ((df[true_col] == 0) & (df[pred_col] == 0)).sum()
fp = ((df[true_col] == 0) & (df[pred_col] == 1)).sum()
fn = ((df[true_col] == 1) & (df[pred_col] == 0)).sum()

print("TP:", tp)
print("TN:", tn)
print("FP:", fp)
print("FN:", fn)
```

```python
# Q3.2: Compute accuracy, precision, recall [We already write this ans]
accuracy = (tp + tn) / (tp + tn + fp + fn) if (tp + tn + fp + fn) > 0 else 0

precision = tp / (tp + fp) if (tp + fp) > 0 else 0
recall = tp / (tp + fn) if (tp + fn) > 0 else 0

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
```

### Q4. Standardization and Min max scaling

Use one numeric column, `monthly_income`.


```python
# Q4.1: Choose the numeric column [2 marks]
income = df['monthly_income']
```

```python
# Q4.2: Standardization with z-score [10 marks]
mean = income.mean()
std = income.std()

z_score = (income - mean) / std

print("Standardized (z-score) values:")
print(z_score.head(10))
print(f"\nMean of standardized values: {z_score.mean():.10f}")
print(f"Std of standardized values: {z_score.std():.10f}")
```

```python
# Q4.3: Min max scaling implementation [10 marks]
min_val = income.min()
max_val = income.max()

min_max = (income - min_val) / (max_val - min_val)

print("Min-Max scaled values:")
print(min_max.head(10))
print(f"\nMin of scaled values: {min_max.min()}")
print(f"Max of scaled values: {min_max.max()}")
```


> **Q4.4 Short answer: [3 marks]**  
> Compare the standardized and min max scaled columns in 2 to 3 sentences.  
> Mention what kind of range each one uses and how the numbers look.

Write your answer here: 

>  Standardization transforms the feature so that the mean becomes 0 and the standard deviation becomes 1, so if you calculate the average after scaling it will be around 0, and the spread will be around 1. Min-max scaling, on the other hand, turns every value into a number between 0 and 1. So the standardized column is centered around zero, while the min-max scaled column stays within the 0–1 range.

### Q5. One hot and ordinal encoding

We will use:
- `city_type` as a nominal feature  
- `satisfaction_level` as an ordinal feature with order `Low` < `Medium` < `High`  


```python
# Q5.1: One hot encoding for city_type using pandas [10 marks]
d_city = pd.get_dummies(df['city_type'], prefix='C', dtype=int)
d_city
```

```python
# Q5.2: Attach one hot encoded columns to df [5 marks]
df = pd.concat([df, d_city], axis=1)

print("One-hot encoded columns added to dataframe:")
print(df[['city_type', 'C_Rural', 'C_Suburban', 'C_Urban']].head())
```

```python
# Q5.3: Ordinal encoding for satisfaction_level [10 marks]
order = {"Low": 1, "Medium": 2, "High": 3}
df['satisfaction_level'] = df['satisfaction_level'].map(order).astype(int)

print("Ordinal encoded satisfaction_level:")
print(df['satisfaction_level'].head(10))
print(f"\nValue counts:")
print(df['satisfaction_level'].value_counts().sort_index())
```


> **Q5.4 Short answer: [5 marks]**  
> In 2 to 3 sentences, explain why one hot encoding is suitable for `city_type`  
> and why ordinal encoding is suitable for `satisfaction_level`.

Write your answer here: .

>  One-hot encoding is suitable for city_type because the categories don’t have any natural order, so each city should get its own separate column
>   For satisfaction_level, ordinal encoding works better because the values have a clear order (like low, medium, high), so we can replace them with numbers that follow that order

### Q6. Euclidean and Manhattan distance

Build 2D vectors for user 0 and user 1 using:
- `income_std`  
- `daily_app_opens` (or its min max scaled version if you prefer)


```python
# Q6.1: Build 2D vectors for first two users [We already write this ans]
vec_cols = ["monthly_income", "daily_app_opens"]

v1 = df.loc[0, vec_cols].values
v2 = df.loc[1, vec_cols].values

print("v1:", v1)
print("v2:", v2)
```

```python
# Q6.2: Euclidean distance computation [5 marks]
eu = np.linalg.norm(v1-v2)
print("Euclidean:", eu)
```

```python
# Q6.3: Manhattan distance computation [5 marks]
ma = np.linalg.norm(v1-v2, ord=1)
print("Manhattan:", ma)
```
