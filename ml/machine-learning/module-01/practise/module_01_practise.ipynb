```python
import numpy as np
import pandas as pd
df = pd.DataFrame([45, 48, 50, 52, 55, 58, 60, 62, 65, 68, 70, 72, 75, 78, 80, 85, 90, 95, 150, 500])
# df.columns[]
# df = df.set_axis(['column'], axis=1)
df.columns = ['salary']
```

```python
# variance, SD, Q2, IQR
var = df['salary'].var()
new_v = np.var(df['salary'], ddof=1)
print(f"pandas variance {var} numpy varian {new_v}")
```

```python
# std = np.std(df['salary'])
std = df['salary'].std()
median = df['salary'].median()
q1 = df['salary'].quantile(0.25)
q2 = df['salary'].quantile(0.50)
q3 = df['salary'].quantile(0.75)
IQR = q3 - q1
lower_fence = q1 - 1.5*IQR
upper_fence = q3 + 1.5*IQR 


df['is_outlier'] = (df['salary'] < lower_fence) | (df['salary'] > upper_fence)
print(median,"median", std,"standard deviation")
print(f" lower_fence{lower_fence}, higher_fence{upper_fence} \n {IQR}, q1,{q1},q2,{q2},q3,{q3}") 
outliers = df[df['is_outlier']]
outliers
```
