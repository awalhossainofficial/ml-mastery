## Module 01: Descriptive Statistics and Distributions

```python
import numpy as np
import pandas as pd

# Simple dataset (e.g., house prices in thousands)
data = [120, 125, 128, 122, 126, 124, 600]  # 600 is an outlier
df = pd.DataFrame({'HousePrice': data})

mean = df['HousePrice'].mean()
print("Mean:", mean)
```

```python
median_value = df['HousePrice'].median()
print(f"median value: {median_value}")
```

**Categorical Data**
- for categorical data, we use mode
- mode is the value that appears most frequently in the dataset
- it is useful for nominal data
- it is not useful for ordinal data

```python
df['City'] = ['Dhaka', 'Dhaka', 'Chattogram', 'Sylhet', 'Dhaka', 'Rajshahi', 'Chattogram']
mode_value = df['City'].mode()[0]
print("mode_value:",mode_value)
```

```python

import numpy as np
import pandas as pd

# Monthly customer purchase amounts
data = [120, 130, 125, 140, 135, 200, 220]
df = pd.DataFrame({'Purchase': data})

mean = np.mean(df['Purchase'])
pop_var = np.var(df['Purchase'], ddof=0)
sam_var = np.var(df['Purchase'], ddof=1)
pop_sd= np.std(df['Purchase'],ddof=0) # population SD
sam_sd = np.std(df['Purchase'],ddof=1) # Sample SD

print('Population varianc',pop_var)
print('Sample varianc',sam_var)
print('Population varianc',pop_sd)
print('Population varianc',sam_sd)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['Purchase'], kde=True, label="smo")
# plt.xlabel("soe")
plt.axvline(mean, color='red', linestyle='--', label='Mean')
plt.legend()
```

```python
import numpy as np
data = [10, 12, 13, 15, 18, 21, 23, 25, 28, 30, 34, 35, 37, 40]

print(f"25th percentile (Q1):",np.percentile(data,25))
print(f"50th percentile (Q1):",np.percentile(data,50))
print(f"50th percentile (Q1):",np.percentile(data,70))
```

```python
q1 = np.percentile(data,25)
q3 = np.percentile(data, 75)
IQR = q3 - q1
lower_bound = q1 - 1.5 * IQR
higher_bound = q3 + 1.5*IQR

print("IQR", IQR)
print(f"Outlier Range {lower_bound} to {higher_bound}")
```

```python
import numpy as np

data = np.array([10, 12, 13, 15, 18, 21, 23, 25, 28, 30, 34, 35, 37, 40])
mean = np.mean(data)
std = np.std(data)

z_score = [(x-mean)/std for x in data]
print("Z-scores: ", np.round(z_score))
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'Income': [22, 25, 27, 29, 35, 40, 42, 100, 110, 115]})

# IQR পদ্ধতি
Q1 = df['Income'].quantile(0.25)
Q3 = df['Income'].quantile(0.75)
IQR = Q3 - Q1

mean = df['Income'].mean()
std = df['Income'].std()
df['z_score'] = (df['Income'] - mean) / std
outliers_z = df[np.abs(df['z_score']) > 2.5]
print("Outliers using Z-score:\n", outliers_z)


# Modified Z-score (Robust)
median = df['Income'].median()
mad = np.median(np.abs(df['Income'] - median))
df['mod_z'] = 0.6745 * (df['Income'] - median) / mad
outliers_mz = df[np.abs(df['mod_z']) > 3.5]
print("Outliers using Mdoified z-score: \n", outliers_mz)
```
