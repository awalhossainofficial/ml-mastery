```python
inpt = input().split()
count_a = 0
count_b = 0
prediction_length = len(inpt)

for i in inpt:
    if i == "A":
        count_a +=1
    elif i == "B":
        count_b +=1

if count_a / prediction_length > 0.7 or count_b / prediction_length > 0.7:
    print("Biased Model")
else:
    print("Fair Model")
```
