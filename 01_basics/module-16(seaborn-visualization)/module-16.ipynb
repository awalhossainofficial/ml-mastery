## Load SNS Data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
student = pd.read_csv('student.csv')
```

## lineplot using seaborn

```python
sns.lineplot(data=student, x='week', y='attendance_rate', errorbar=None, hue='gender')
```

```python
# Figure Level
sns.relplot(kind='line', data=student, x='week', y='attendance_rate', errorbar=None, hue='gender')
```

## Scatter Plot using seaborn

```python
sns.scatterplot(data=student, x='study_hours', y='test_score',hue='gender', style='class_level')
```

```python
# now let's try with figure level
sns.relplot(kind='scatter',data=student, x='study_hours',y='test_score',hue='gender', style='class_level',size='Tshirt_size')
```

```python
tips_data = sns.load_dataset("tips")

tips_data.head() 
```

```python

sns.relplot(kind ='scatter' , data = tips_data , x = 'total_bill' , y = 'tip',hue='sex',style='time',size='size', col='time')
```

## **Facet Plot using seaborne**

```python
sns.relplot(kind='scatter',data=student,x='study_hours',y='test_score',col='week',col_wrap=2)
```

## **Histogram using seaborn**

```python
sns.histplot(data=student, x='attendance_rate', hue='gender',element='poly',kde=True, stat='count')
```

## **KDE plot using seaborn**

```python
sns.displot(kind='hist', data=tips_data, x='total_bill',col='sex')
tips_data
```

## **KDE PLot**

```python
sns.kdeplot(data=tips_data, x="tip", hue='sex', fill=True)
```

## **Count Plot & Bar Chart**

```python
sns.countplot(data=tips_data, x="day", order=["Thur", "Fri", "Sat", "Sun"])
# plt.show()
```

## Bar plot


```python
sns.barplot(data=student, x='gender', y='test_score', errorbar=None, estimator=np.median)
```

## RegPlot & Implot

```python
sns.regplot(data=student, x='study_hours', y='test_score')
```

```python
## figure Level 
sns.lmplot(data=student, x='study_hours', y='test_score', hue='gender')
```

## Pairplot & Joinplot

```python
stud_marks = student[['Marks', 'study_hours', 'attendance_rate','gender']]
sns.pairplot(data=stud_marks, hue='gender')
```

```python
sns.pairplot(data=stud_marks, kind='hist')
```

```python
#join-plot
sns.jointplot(data=student, x='study_hours', y='Marks', hue='gender')
```

```python
sns.jointplot(data=stud_marks , x ='study_hours' , y='Marks',kind='kde') 
```

## **Plotly- An interactive library for creating plots**

```python
fig = px.scatter(student, x='time_study', y='Marks', color='gender',size='Tshirt_size', hover_data=['hostel'])
fig.show()
```

```python
#lineplot using plotly

enrollment = pd.read_csv('enrollment.csv')

fig = px.line(enrollment, x='Year', y='Programming', markers=True)
fig.show()
```

```python
#histogram
fig = px.histogram(stud_marks, x='attendance_rate',color='gender',nbins=10,barmode='overlay')
fig.show()
```
